## Exercise 2

Suppose we have 50 points between 0 and 100 to be spread randomly. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import random
import numpy as np
from mpl_toolkits import mplot3d
from sklearn.neighbors import NearestNeighbors


%matplotlib inline
```

### First try with one dimension

`1.`Generate and plot the 50 points within 5 intervals:  
- [0,20), 
- [20,40), 
- [40,60), 
- [60,80),
- [80,100)

```python
# Unique random numbers are generated
n = random.sample(range(0,100), 50)

# Intervals assignment
intervals = {1 : [0,20], 2 : [20,40], 3 : [40,60],
             4 : [60,80], 5 : [80,100]}
   
assignment_1D = {}
for i in [1,2,3,4,5]:
    assignment_1D[i] = []       
        
for number in n:
    
    for i in [1,2,3,4,5]:
        
        start = intervals[i][0]
        end = intervals[i][1]
        
        if number >= start and number < end :
            assignment_1D[i].append(number)
            
```

```python
# Numbers assigned to their intervals
assignment_1D
```

#### Plot

```python
# Different colors and markers for each interval

number_of_markers = 5
markers = ['o','s','^','P','p','.','v','>','<','1','2','3','4','8','*','h','H','+','x','X','d','D']
markers = random.choices(markers,k=number_of_markers)
markers = np.reshape(markers , (5,1))

number_of_colors = 5
colors = ["#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)])
             for i in range(number_of_colors)]
colors = np.reshape(colors, (5,1))




# Plot
fig = plt.figure(figsize=(18,1))

for i in [1,2,3,4,5]:
    plt.scatter(assignment_1D[i],np.zeros(np.shape(assignment_1D[i])[0]), marker = markers[i-1][0], color = colors[i-1][0])
    
    
plt.title('1D random number')
plt.grid(color = 'gray', linestyle = '-', linewidth = 0.5)
plt.show()
```

## Now let's try with 2 dimensions

`2.`Assuming we have 5 intervals on the first dimension and 5 intervals on the 2nd dimension, generate 50 points within these intervals and plot them.

```python
# Unique random numbers are generated
n = random.sample(
    [[x,y] for x in range(100) for y in range(100)]
    , 50)

# Intervals assignment
intervals_x = {1 : [0,20], 2 : [20,40], 3 : [40,60],
               4 : [60,80], 5 : [80,100]}
intervals_y = {1 : [0,20], 2 : [20,40], 3 : [40,60],
               4 : [60,80], 5 : [80,100]}


assignment_2D = {}
for i in [1,2,3,4,5]:
    assignment_2D[i] = {}
    for j in [1,2,3,4,5]:
        assignment_2D[i][j] = []


for number in n:
    
    for i in [1,2,3,4,5]:
        for j in [1,2,3,4,5]:
            
            start_x = intervals_x[i][0]
            end_x   = intervals_x[i][1]
            
            start_y = intervals_y[j][0]
            end_y   = intervals_y[j][1]
        
        
            if number[0] >= start_x and number[0] < end_x :
                if number[1] >= start_y and number[1] < end_y :
                    assignment_2D[i][j].append(number)
                    
```

## Now in 3D Case

`3.`Now suppose we have 5 intervals on the third dimension.With this condition, generate 50 points within them and plot the points in a 3d manner.

```python
# Unique random numbers are generated
n = random.sample(
    [[x,y,z] for x in range(100) for y in range(100) for z in range(100)]
    , 50)

# Intervals assignment
intervals_x = {1 : [0,20], 2 : [20,40], 3 : [40,60],
               4 : [60,80], 5 : [80,100]}
intervals_y = {1 : [0,20], 2 : [20,40], 3 : [40,60],
               4 : [60,80], 5 : [80,100]}
intervals_z = {1 : [0,20], 2 : [20,40], 3 : [40,60],
               4 : [60,80], 5 : [80,100]}

assignment_3D = {}
for i in [1,2,3,4,5]:
    assignment_3D[i] = {}
    for j in [1,2,3,4,5]:
        assignment_3D[i][j] = {}
        for k in [1,2,3,4,5]:
            assignment_3D[i][j][k] = []

for number in n:
    
    for i in [1,2,3,4,5]:
        for j in [1,2,3,4,5]:
            for k in [1,2,3,4,5]:
                
                start_x = intervals_x[i][0]
                end_x   = intervals_x[i][1]
                
                start_y = intervals_y[j][0]
                end_y   = intervals_y[j][1]
                
                start_z = intervals_z[k][0]
                end_z   = intervals_z[k][1]
                
                if number[0] >= start_x and number[0] < end_x :
                    if number[1] >= start_y and number[1] < end_y :
                        if number[2] >= start_z and number[2] < end_z :
                            assignment_3D[i][j][k].append(number)
                                   
```

### $F_i$

Considering :

$$
F_i = \frac{f_i}{N}
$$

Where $N$ is the number of samples and $f_i$ is the frequency of data in the given intervals, meaning how many samples are in a specific interval.

intervals are : 

- [0,20)

- [20, 40)

- [40, 60)

- [60, 80)

- [80, 100)

for all xyz axis.

```python
n = 50

# 1D

F_1D = {}
for i in [1,2,3,4,5]:
    F_1D[i] = len(assignment_1D[i])/n


    
# 2D

F_2D = {}
for i in [1,2,3,4,5]:
    F_2D[i] = {}
    for j in [1,2,3,4,5]:
        F_2D[i][j] = len(assignment_2D[i][j])/n

# 3D 

F_3D = {}
for i in [1,2,3,4,5]:
    F_3D[i] = {}
    for j in [1,2,3,4,5]:
        F_3D[i][j] = {}
        for k in [1,2,3,4,5]:
            F_3D[i][j][k] = len(assignment_3D[i][j][k])/n    
```

#### Print the results :


- 1 mean [0, 20)

- 2 means [20, 40)

- 3 means [40, 60)

- 4 means [60, 80)

- 5 means [80, 100)

```python
dic = {1 : '[0,20)', 2 : '[20,40)', 3 : '[40,60)', 4 : '[60,80)', 5 : '[80,100)'}

print("####################")
print("1D Freq/N of each cell: ")
for F in F_1D:
    print(f"F in {dic[F]} : ", F_1D[F])
print("####################")


print("####################")
print("1D Freq/N of each cell: ")
for i in F_2D:
    for F in F_2D[i]:
        print(f"F in {dic[i]}{dic[F]} : ", F_2D[i][F])
    
print("####################")

print("####################")
print("3D Freq/N of each cell: ")
for i in F_3D:
    for j in F_3D[i]:
        for F in F_3D[i][j]:
            print(f"F in {dic[i]}{dic[j]}{dic[F]} : ", F_3D[i][j][F])
    
print("####################")
```

**Answer:**



F is gradually decreasing, meaning that the number of samples in each cell is lower that what it was previously.
There are also many cells that contain no sample. To fix this problem, we either have to augment our data by generating more random numbers or use other methods to get rid of extra dimensions (PCA , ...).


Theoretically:

$\bullet$ In 1D case, we have 3 cells and 50 samples which means at best, each cell should contain $\frac{50}{3} = 16.7$ samples.


$\bullet$ In 2D case, we have 25 cells and 50 samples which means at best, each cell should contain $\frac{50}{25} = 2$ samples.

$\bullet$ In 3D case, we have 125 cells and 50 samples which means at best, 50 cells should contain 1 sample and 75 cells are empty.


### Can we generalize to more than three dimensions? 

Let's draw a bunch of random points from a 0-1 distribution, then measure how far away from the origin they are. (4D to 10D)


```python
def norm(x):
    return np.linalg.norm(x,2)

def inball(x, R):
    if norm(x) <= R:
        return 1
    return 0

def what_percent_of_the_ncube_is_in_the_nball(samples, R ,sample_size=10**4):
    sum = 0
    for sample in samples:
        if inball(sample, R):
            sum+=1
    return sum/sample_size
```

#### Generalized result :

2D to 14D random numbers are generated and their frequency is calculated, then reported and visualized.

```python
# N : number of samples, R : radius of the circle

N  = 1000
R  = 1

Fre = []
# Generate random numbers

for i in range(2,15):
    samples = np.random.uniform(-2*R,2*R,(N,i))
    
    # Calculate f
    f = what_percent_of_the_ncube_is_in_the_nball(samples, R, N)
    Fre.append(f)
    print(f'Frequency/N of samples within the {i}d-ball is :',f) 
    
plt.scatter(range(2,15),Fre,marker = 's', color = 'red')
plt.plot(range(2,15),Fre,linestyle=':', color = 'blue')
plt.xlabel('Dimension')
plt.ylabel('Frequency/N')
plt.title('Frequency vs dimension plot')
```

## Exercise 4

`1.`Generate 50 points randomly with a uniform distribution in the range of 0 to 100 for one, three, ten and one hundred dimensions, then calculate the average distance of 5 nearest neighbors (5-NN) for 10 random points in each dimension.
Discuss your answer.

```python
# Generate 50 random numbers
N = 50
n_neighbors_ = 5


samples_d = {}
points_d  = {}
mean_d    = {}


# For each dimension

for d in [1,3,10,100]:
    samples_d[d] = np.random.uniform(0,100,(N, d))
    # 10 random point within the sample
    points_d[d] = random.choices(samples_d[d], k = 10)
    
    # make the model
    
    KNN_l2 = NearestNeighbors(n_neighbors = n_neighbors_, metric='minkowski', p= 2)
    KNN_l2.fit(samples_d[d])
    
    KNN_l1 = NearestNeighbors(n_neighbors = n_neighbors_, metric='minkowski', p= 1)
    KNN_l1.fit(samples_d[d])
    
    KNN_cos = NearestNeighbors(n_neighbors = n_neighbors_, metric='cosine', p= 2)
    KNN_cos.fit(samples_d[d])
    
    # calculate 5-NN for the point and their distance mean
    
    mmeans = {}
    
    # l2 
    means = []
    distances = KNN_l2.kneighbors(points_d[d])[0]
    for i in range(10):
        means.append(distances[i].mean())
    mmeans['l2'] = means
    print(f'mean distance of 10 points using l2 5-NN ({d}D samples) : ', means)
    print('\n')
    
    # l1
    means = []
    distances = KNN_l1.kneighbors(points_d[d])[0]
    for i in range(10):
        means.append(distances[i].mean())
    mmeans['l1'] = means
    print(f'mean distance of 10 points using l1 5-NN ({d}D samples) : ', means)
    print('\n')
    
    # cosine
    means = []
    distances = KNN_cos.kneighbors(points_d[d])[0]
    for i in range(10):
        means.append(distances[i].mean())
    mmeans['cosine'] = means
    print(f'mean distance of 10 points using cosine 5-NN ({d}D samples) : ', means)
    print('\n')
    
    
    print('\n##########################\n')
    
    mean_d[d] = mmeans
    
```

#### Visualize the result


```python
fig = plt.figure(figsize=(25,10))


# 1D
ax = fig.add_subplot(1,4,1)
ax.scatter(range(1,11),mean_d[1]['l2'])
ax.plot(range(1,11),mean_d[1]['l2'], label = '1D - L2', alpha = 0.7)

ax.scatter(range(1,11),mean_d[1]['l1'])
ax.plot(range(1,11),mean_d[1]['l1'], label = '1D - L1', alpha = 0.5)

ax.scatter(range(1,11),mean_d[1]['cosine'])
ax.plot(range(1,11),mean_d[1]['cosine'], label = '1D - Cosine', alpha = 0.4)

ax.set_ylabel('mean')
ax.set_xlabel('index')
ax.set_title('1D')
ax.legend(loc = 'best')

# 3D
ax = fig.add_subplot(1,4,2)
ax.scatter(range(1,11),mean_d[3]['l2'])
ax.plot(range(1,11),mean_d[3]['l2'], label = '3D - L2', alpha = 0.7)

ax.scatter(range(1,11),mean_d[3]['l1'])
ax.plot(range(1,11),mean_d[3]['l1'], label = '3D - L1', alpha = 0.5)

ax.scatter(range(1,11),mean_d[3]['cosine'])
ax.plot(range(1,11),mean_d[3]['cosine'], label = '3D - Cosine', alpha = 0.4)

ax.set_ylabel('mean')
ax.set_xlabel('index')
ax.set_title('3D')
ax.legend(loc = 'best')

# 10D
ax = fig.add_subplot(1,4,3)
ax.scatter(range(1,11),mean_d[10]['l2'])
ax.plot(range(1,11),mean_d[10]['l2'], label = '10D - L2', alpha = 0.7)

ax.scatter(range(1,11),mean_d[10]['l1'])
ax.plot(range(1,11),mean_d[10]['l1'], label = '10D - L1', alpha = 0.5)

ax.scatter(range(1,11),mean_d[10]['cosine'])
ax.plot(range(1,11),mean_d[10]['cosine'], label = '10D - Cosine', alpha = 0.4)

ax.set_ylabel('mean')
ax.set_xlabel('index')
ax.set_title('10D')
ax.legend(loc = 'best')

# 100D
ax = fig.add_subplot(1,4,4)
ax.scatter(range(1,11),mean_d[100]['l2'])
ax.plot(range(1,11),mean_d[100]['l2'], label = '100D - L2', alpha = 0.7)

ax.scatter(range(1,11),mean_d[100]['l1'])
ax.plot(range(1,11),mean_d[100]['l1'], label = '100D - L1', alpha = 0.5)

ax.scatter(range(1,11),mean_d[100]['cosine'])
ax.plot(range(1,11),mean_d[100]['cosine'], label = '100D - Cosine', alpha = 0.4)

ax.set_ylabel('mean')
ax.set_xlabel('index')
ax.set_title('10D')
ax.legend(loc = 'best')


fig.tight_layout()
plt.show()
```

**Answer :**

We can see that cosine metric works best in almost every case, therefore it is possible to remedy curse of dimensionality using proper metrics. 

Other methods to remedy dimensionality are using feature selection methods or data augmentation.
