Just as the title says. Import the necessary libraries.

```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
```

## <font color='#1111cd'>   Many useless columns and new dataframes

    
Since our objective only requires MaxTemp and MinTemp, then all the other columns are irrelavent. Our objective is to find a relationship between min and max temperature.
    
It is obvious that many features in these datasets affect the max and min temperature, However our objective is a function that can predict max temperature given min temperature.

```python
task = dataset[['MaxTemp','MinTemp']].sort_values(by = 'MinTemp')
```

```python
task.describe()
```

## <font color='#1111ef'>   train and test split
    
Split a portion of data so that it can be used as test to evaluate our model. 

```python
task_x = task['MinTemp']
task_y = task['MaxTemp']
task_x_train, task_x_test, task_y_train, task_y_test = train_test_split(task_x, task_y, test_size= 0.25)
```

# <font color='#1111ef'>   Visualization
    
<hr>
    
Visualizing task on  XY-plot. MinTemp relation with  MaxTemp :
    
As you can see it has a trend , and probably a cycle which is causing it to look like this. We ignore such details since our objective is much simpler.

```python
plt.plot(task_x,task_y,color = "red",label = 'plot')
plt.scatter(task_x,task_y,color = "blue",marker = '.', label = 'scatter')
plt.xlabel('MinTemp')
plt.ylabel('MaxTemp')
plt.legend(loc = 'best')
plt.show()
```

# <font color='#1111ef'>   Model
 
<hr>
    
    
Data is ready to be used by our model, so before that we need to create multiple models so that each model will train on the training data and evaluated on the test data.
    
    
We choose a linear regression and polynomial regression for test tasks. degree of polynomial regressions are 1, 2, 3, 4, 5, 10, 20, 30, 50, 100. Overall 10 linear regression models  for this task.
    
    
As well as linear and polynomial regression we are going to use XGBoost to model our data and evaluate it as well.
    
- XGBoost
    
This is going to take a while to run  (probably) ....

```python
# This might take some time . ....

task_models = {}


M = [1, 2, 3, 4, 5, 10, 20, 30, 50, 100]
polynomials = {}
for _degree in M:
    polynomials[_degree] = PolynomialFeatures(degree = _degree, include_bias = True)

    
# Linear Regression models 

for _degree in M:
        model = LinearRegression()
        
        x_train = task_x_train.values.reshape(-1,1)
        x_train = polynomials[_degree].fit_transform(x_train)
        y_train = task_y_train.values
        
        model.fit(x_train,y_train)
        
        task_models[_degree] = model
        
# XGBOOST


xgb= XGBRegressor(objective ='reg:squarederror',n_estimators=1000, max_depth=7, eta=0.1, subsample=0.7, colsample_bytree=0.8)
xgb.fit(task_x_train,task_y_train)
task_models['XGBoost'] = xgb
```

## <font color='#1111ef'>   Predict
    
Our goal is to use these models to predict the MaxTemp  using MinTemp.

```python
task_predicted = {}

for _degree in M:
    
    task_predicted[_degree] = {
        'train' : task_models[_degree].predict(polynomials[_degree].fit_transform(task_x_train.values.reshape(-1,1))), 
        'test' : task_models[_degree].predict(polynomials[_degree].fit_transform(task_x_test.values.reshape(-1,1)))
        }
    
task_predicted['XGBoost'] = {
        'train' : task_models['XGBoost'].predict(task_x_train),
        'test' : task_models['XGBoost'].predict(task_x_test)   
}
```

## <font color='#1111ef'>   RMSE
    
We are going to use RMSE to evaluate our models; Therefore we need to define it :
    
Edit :  It is also possible to use it from sklearn
    
    
$RMSE = \sqrt{\frac{\sum_{i=1}^{n}(\hat y_i - y_i)^2}{n}}$  

```python
def RMSE(Y,Yhat):
    N = len(Y)
    rmse = np.sqrt(np.sum((Y -Yhat)**2) / N )
    return rmse
```

## <font color='#1111ef'> Evaluation
    
    
Evaluate our models using RMSE

```python
task_rmse = {}

for _degree in M: 
    task_rmse[_degree]  = {
        'train' : RMSE(task_y_train,task_predicted[_degree]['train']),
        'test' : RMSE(task_y_test,task_predicted[_degree]['test'])
    }
    
task_rmse['XGBoost'] = {
        'train' : RMSE(task_y_train,task_predicted['XGBoost']['train']),
        'test' : RMSE(task_y_test,task_predicted['XGBoost']['test'])
}
 
```

# <font color='#1111ef'>   Result
 
<hr>
    
You can check the result for each model and choose the best model to predict the MaxTemp using MinTemp.

    
We will also Visualize our predicted vs actual data on both train and test for our task.

```python
plt.plot(M, [task_rmse[i]['train'] for i in M], color = 'red', label = 'train')
plt.scatter(M, [task_rmse[i]['train'] for i in M], color = 'red', marker= 's')
plt.plot(M, [task_rmse[i]['test'] for i in M], color = 'blue', label = 'test')
plt.scatter(M, [task_rmse[i]['test'] for i in M], color = 'blue', marker= 's')
plt.plot(M, [task_rmse['XGBoost']['train'] for i in M],color = 'lime' 
         , linestyle = 'dotted',label = 'XGBoost - train',alpha = 0.5)
plt.plot(M, [task_rmse['XGBoost']['train'] for i in M],color = 'purple' 
         , linestyle = 'dashed',label = 'XGBoost - test',alpha = 0.5)
plt.xlabel('M')
plt.ylabel('RMSE')
plt.title('Task')
plt.legend(loc = 'best')
plt.show()
```

## <font color='#1111ef'>   Linear/Polynomials Regression Models
